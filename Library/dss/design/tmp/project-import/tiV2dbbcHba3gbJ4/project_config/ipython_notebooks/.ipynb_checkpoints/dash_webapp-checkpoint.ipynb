```python
import dataiku
from dataiku import pandasutils as pdu
import pandas as pd
```

```python
# Example: load a DSS dataset as a Pandas dataframe
mydataset = dataiku.Dataset("monthly_data_py_prep")
df = mydataset.get_dataframe()
df.head()
```

```python

import plotly.figure_factory as ff

import pandas as pd
```

```python
s_df = df.pivot_table(values=['WO Ees', 'WO DEALS', 'DC Total', 'WO BP Total'], index=['DM_NAME', 'REGION'])
s_df['dc_meas'] = s_df['DC Total'].apply(lambda x: [round(x,2), round(x + 5,2)])
s_df['bp_meas'] = s_df['WO BP Total'].apply(lambda x: [round(x, 2), round(x + 5, 2)])
s_df['wo_ees_meas'] = s_df['WO Ees'].apply(lambda x: [round(x, 2), round(x + 5,2)])
s_df['wo_deal_meas'] = s_df['WO DEALS'].apply(lambda x: [round(x, 2), round(x + 5,2)])

s_df['dc_mark'] = s_df['DC Total'].max()
s_df['bp_mark'] = s_df['WO BP Total'].max()
s_df['wo_ee_mark'] = s_df['WO Ees'].max()
s_df['wo_deal_mark'] = s_df['WO DEALS'].max()



s_df['dc_range'] = s_df['DC Total'].apply(lambda x: [round(s_df['DC Total'].quantile(.25),2), 
                    round(s_df['DC Total'].quantile(.5),2), round(s_df['DC Total'].quantile(.75),2)])
s_df['bp_range'] = s_df['WO BP Total'].apply(lambda x: [round(s_df['WO BP Total'].quantile(.25),2), 
                    round(s_df['WO BP Total'].quantile(.5),2), round(s_df['WO BP Total'].quantile(.75),2)])
s_df['wo_ees_range'] = s_df['WO Ees'].apply(lambda x: [round(s_df['WO Ees'].quantile(.25),2), 
                    round(s_df['WO Ees'].quantile(.5),2), round(s_df['WO Ees'].quantile(.75),2)])
s_df['wo_deals_range'] = s_df['WO DEALS'].apply(lambda x: [round(s_df['WO DEALS'].quantile(.25),2), 
                    round(s_df['WO DEALS'].quantile(.5),2), round(s_df['WO DEALS'].quantile(.75),2)])

s_df = s_df.drop(['DC Total', 'WO BP Total', 'WO Ees', 'WO DEALS'], axis=1)
for x in ['dc_mark', 'bp_mark', 'wo_ee_mark', 'wo_deal_mark']:
    s_df[x] = s_df[x].apply(lambda x: [x])
s_df.reset_index(level=0, inplace=True)
s_df.reset_index(level=0, inplace=True)
rep_names = [x for x in s_df.DM_NAME.unique()]
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']


app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

app.layout = html.Div(children=[
    html.H1(children='Insperity Scorecard'),

    html.Label(children='Region'),

    dcc.RadioItems(
        id='region-dropdown',
        options=[{'label': k, 'value': k} for k in s_df.REGION.unique()],
        value='Eastern'
    ),
    html.Hr(),

    html.Label('Sales Reps'),

    dcc.Dropdown(id='reps_by_region'),
    html.Hr(),

    html.H3(children='WO Deals'),
    dcc.Graph(id='wo-deal'),
    html.H3(children='WO EEs'),
    dcc.Graph(id='wo-ees'),
    html.H3(children='DCs'),
    dcc.Graph(id='dc'),
    html.H3(children='BPs'),
    dcc.Graph(id='bp')

])

@app.callback(
    dash.dependencies.Output('reps_by_region', 'options'),
    [dash.dependencies.Input('region-dropdown', 'value')])

def set_rep_options(selected_region):
    return [{'label': i, 'value': i} for i in s_df[s_df.REGION == selected_region]['DM_NAME'].unique()]


@app.callback(
    dash.dependencies.Output('reps_by_region', 'value'),
    [dash.dependencies.Input('reps_by_region', 'options')])
def set_rep_value(available_options):
    return available_options[0]['value']

#update first bullet
@app.callback(
    dash.dependencies.Output('wo-deal', 'figure'),
    [dash.dependencies.Input('reps_by_region', 'value')])

def build_graph_one(Rep):
    plot_df = s_df[s_df['DM_NAME'] == Rep]
    return ff.create_bullet(plot_df, markers='wo_deal_mark', measures='wo_deal_meas',
            ranges='wo_deals_range', titles="DM_NAME")

#update second bullet
@app.callback(
    dash.dependencies.Output('wo-ees', 'figure'),
    [dash.dependencies.Input('reps_by_region', 'value')])

def build_graph_two(Rep):
    plot_df = s_df[s_df['DM_NAME'] == Rep]
    return ff.create_bullet(plot_df, markers='wo_ee_mark', measures='wo_ees_meas',
            ranges='wo_ees_range', titles="DM_NAME")

#update third bullet
@app.callback(
    dash.dependencies.Output('dc', 'figure'),
    [dash.dependencies.Input('reps_by_region', 'value')])

def build_graph_three(Rep):
    plot_df = s_df[s_df['DM_NAME'] == Rep]
    return ff.create_bullet(plot_df, markers='dc_mark', measures='dc_meas',
            ranges='dc_range', titles="DM_NAME")


#update four bullet
@app.callback(
    dash.dependencies.Output('bp', 'figure'),
    [dash.dependencies.Input('reps_by_region', 'value')])

def build_graph_four(Rep):
    plot_df = s_df[s_df['DM_NAME'] == Rep]
    return ff.create_bullet(plot_df, markers='bp_mark', measures='bp_meas',
            ranges='bp_range', titles="DM_NAME")



if __name__ == '__main__':
    app.run_server(debug=False, threaded=False, port=8051)
```
